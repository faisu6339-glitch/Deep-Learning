```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

# Black image
img = np.zeros((400,400,3), dtype=np.uint8)
```

#Draw Rectangle

```python
cv2.rectangle(img, (50,50), (200,200), (0,255,0), 3)
```

#Draw Circle

```python
cv2.circle(img, (300,100), 50, (255,0,0), -1)
```

#Draw Line

```python
cv2.line(img, (0,300), (400,300), (0,0,255), 5)
```

#Add Text

```python
cv2.putText(img, "OpenCV", (100,380),
            cv2.FONT_HERSHEY_SIMPLEX,
            1, (255,255,255), 2)
```

```python
cv2_imshow(img)
```

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

img = np.zeros((400,400,3), dtype=np.uint8)

cv2.rectangle(img, (50,50), (200,200), (0,255,0), 3)
cv2.circle(img, (300,100), 50, (255,0,0), -1)
cv2.line(img, (0,300), (400,300), (0,0,255), 5)

pts = np.array([[250,250],[300,350],[200,350]], np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(img, [pts], True, (255,255,0), 3)

cv2.putText(img, "OpenCV", (100,380),
            cv2.FONT_HERSHEY_SIMPLEX,
            1, (255,255,255), 2)

cv2_imshow(img)
```
