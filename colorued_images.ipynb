```python
import numpy as np
import matplotlib.pyplot as plt

# Red image
red = np.zeros((100,100,3), dtype=np.uint8)
red[:,:,0] = 255   # Red channel

# Green image
green = np.zeros((100,100,3), dtype=np.uint8)
green[:,:,1] = 255  # Green channel

# Blue image
blue = np.zeros((100,100,3), dtype=np.uint8)
blue[:,:,2] = 255   # Blue channel

# Show images
plt.imshow(red)
plt.title("Red")
plt.show()

plt.imshow(green)
plt.title("Green")
plt.show()

plt.imshow(blue)
plt.title("Blue")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((256, 256, 3), dtype=np.uint8)

for i in range(256):
    for j in range(256):
        img[i, j] = [i, j, (i+j)//2]  # R=i, G=j, B=mix

plt.imshow(img)
plt.title("All Colors Spectrum")
plt.axis('off')
plt.show()
```

```python
img = np.zeros((256, 256, 3), dtype=np.uint8)

img[:,:,0] = np.arange(256).reshape(256,1)   # Red varies vertically
img[:,:,1] = np.arange(256).reshape(1,256)   # Green varies horizontally
img[:,:,2] = 128  # Constant blue

plt.imshow(img)
plt.title("RGB Color Space")
plt.axis('off')
plt.show()
```

```python
img = np.zeros((256, 256, 3), dtype=np.uint8)

for i in range(256):
    for j in range(256):
        img[i,j] = [i, j, 255-j]

plt.imshow(img)
plt.title("Full RGB Variation")
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create empty image
img = np.zeros((256, 256, 3), dtype=np.uint8)

# Fill colors
for i in range(256):        # rows → Red
    for j in range(256):    # columns → Green
        img[i, j] = [i, j, 128]   # Blue fixed

# Show image
plt.imshow(img)
plt.title("256x256 Color Space (R vs G)")
plt.axis('off')
plt.show()
```
