#Chess board

```python
zeros=np.zeros((8,8),dtype=int)
```

```python
for i in range(len(zeros)):
  for j in range(len(zeros[0])):
    if (i+j)%2==0:
      zeros[i][j]=1
plt.imshow(zeros)
```

```python
ones=np.ones((8,8),dtype=int)
for i in range(len(ones)):
  for j in range(len(ones[i])):
    if (i+j)%2==0:
      ones[i][j]=0
plt.imshow(ones)
```
