```python
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((100,100,3), dtype=np.uint8)

img[:,:,0] = 255   # Red (max)
img[:,:,1] = 150   # Green (adds lightness)
img[:,:,2] = 150   # Blue (adds lightness)

plt.imshow(img)
plt.title("Light Red")
plt.axis('off')
plt.show()
```

```python
img = np.zeros((256,256,3), dtype=np.uint8)

for i in range(256):
    img[:,i] = [255, i, i]   # Red fixed, G & B increase

plt.imshow(img)
plt.title("Light Red Gradient")
plt.axis('off')
plt.show()
```

#Custom RGB Colors

```python
import numpy as np
import matplotlib.pyplot as plt

arr = np.array([
    [[255,0,0],   [0,255,0],   [0,0,255]],
    [[255,255,0], [0,255,255], [255,0,255]],
    [[128,128,128],[255,128,0],[128,0,255]]
], dtype=np.uint8)

plt.imshow(arr)
plt.title("Custom RGB Colors")
plt.axis('off')
plt.show()
```

```python
img = np.zeros((256,256,3), dtype=np.uint8)

for i in range(256):
    for j in range(256):
        img[i,j] = [255, i, j]   # Mix Red + Green + Blue

plt.imshow(img)
plt.title("Custom Color Mix")
plt.axis('off')
plt.show()
```

```python
r = 255
g = np.arange(256).reshape(256,1)
b = np.arange(256).reshape(1,256)

img = np.zeros((256,256,3), dtype=np.uint8)
img[:,:,0] = r
img[:,:,1] = g
img[:,:,2] = b

plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((50, 256, 3), dtype=np.uint8)

for i in range(256):
    img[:, i] = [i, 0, 0]   # Red increases

plt.imshow(img)
plt.title("Shades of Red (Dark → Bright)")
plt.axis('off')
plt.show()
```

#All Red Shades

```python
img = np.zeros((256,256,3), dtype=np.uint8)

for i in range(256):
    for j in range(256):
        img[i,j] = [i, j//2, j//2]

plt.imshow(img)
plt.title("All Red Shades")
plt.axis('off')
plt.show()
```

#Custom Color Grid (Manual Design)

```python
import numpy as np
import matplotlib.pyplot as plt

img = np.array([
    [[255,0,0],   [0,255,0],   [0,0,255],   [255,255,0]],
    [[255,165,0], [255,192,203],[128,0,128],[0,255,255]],
    [[128,128,128],[0,0,0],    [255,255,255],[0,128,128]]
], dtype=np.uint8)

plt.imshow(img)
plt.title("Custom Color Image")
plt.axis('off')
plt.show()
```

```python
img = np.zeros((256,256,3), dtype=np.uint8)

for i in range(256):
    for j in range(256):
        img[i,j] = [i, j, 255-i]   # Custom formula

plt.imshow(img)
plt.title("Custom Pattern Colors")
plt.axis('off')
plt.show()
```

```python
r = np.arange(256).reshape(256,1)
g = np.arange(256).reshape(1,256)

img = np.zeros((256,256,3), dtype=np.uint8)
img[:,:,0] = r          # Red gradient
img[:,:,1] = g          # Green gradient
img[:,:,2] = 255 - r    # Custom blue

plt.imshow(img)
plt.axis('off')
plt.show()
```

#Structured Dense Image

```python
img = np.zeros((256,256,3), dtype=np.uint8)

for i in range(256):
    for j in range(256):
        img[i,j] = [i, j, (i*j)%256]

plt.imshow(img)
plt.title(
    "Dense Gradient Image")
plt.axis('off')
plt.show()
```

```python
img = np.zeros((256,256,3), dtype=np.uint8)

for i in range(256):
    for j in range(256):
        img[i,j] = [(i*5)%256, (j*7)%256, (i+j)%256]

plt.imshow(img)
plt.title("High Frequency Dense Image")
plt.axis('off')
plt.show()
```
