```python
import numpy as np
import matplotlib.pyplot as plt

# Create black and white images
black = np.zeros((256, 256))          # All pixels = 0 (black)
white = np.ones((256, 256)) * 255     # All pixels = 255 (white)

# Display images
plt.figure(figsize=(8,4))

# Black image
plt.subplot(1,2,1)
plt.imshow(black, cmap='gray')
plt.title("Black Image")
plt.axis('off')

# White image
plt.subplot(1,2,2)
plt.imshow(white, cmap='gray')
plt.title("White Image")
plt.axis('off')

plt.show()
```
